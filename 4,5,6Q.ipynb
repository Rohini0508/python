```python
# query
query = """
SELECT
   *
FROM customers;
"""


css = pd.read_sql(query, con=engine)
#css-from sql server to the dataframe(vs code) 

# preview
print(css.head())
```

```python
css['name'] = css['name'].str.replace(r'^(Mr\.|Mrs\.|Miss|Ms\.|Dr\.)\s*', '', regex=True)
css['name'] = css['name'].str.replace(r'\s*(Jr\.|Sr\.|II|III|DDS|PhD|MD|MBA|Esq\.?)$', '', regex=True)
```

```python
css[['first_name', 'last_name']] = css['name'].str.strip().str.split(' ', n=1, expand=True)
print(css[['first_name', 'last_name']])
css
```

```python
# query
queryy = """
SELECT
   *
FROM country_codes;
"""
ccc = pd.read_sql(queryy, con=engine)

# preview
print(ccc.head())
```

```python
css['cntry_code'] = css['address'].str.slice(-8, -6)
```

```python
css
```

```python
ccc
```

```python
merged = css.merge(
    ccc[['Country_code', 'International_dialing']],  
    left_on='cntry_code',     
    right_on='Country_code',    
    how='left'                 
)
```

```python
merged['International_dialing'] = (
    merged['International_dialing']
    .astype(str)
    .str.split(r'\s+(or|and)\s+', expand=False)
    .str[0]
    .str.strip()
)
merged
```

```python
merged['phone'] = merged['phone'].astype(str)
```

```python
merged['International_dialing'] = (
    merged['International_dialing']
    .astype(str)
    .str.replace(r'[^0-9+-]', '', regex=True)
)
merged
```

```python
merged['phone'] = merged['International_dialing'].astype(str) +' '+ merged['phone']
merged
```

```python
merged = merged.drop(columns=['cntry_code'])
merged = merged.drop(columns=['Country_code'])
merged = merged.drop(columns=['International_dialing'])
merged
```
