```python
import pandas as pd
```

```python
customers_upd=pd.read_csv(r"C:\Users\Rohini\Downloads\us_customer_data.csv")
```

```python
transaction_table=pd.read_csv(r"C:\Users\Rohini\Downloads\transaction_data.csv")
```

```python
merged_table=pd.merge(customers_upd,transaction_table,on="customer_id",how="inner")
```

```python
merged_table
```

```python
merged_table.isnull().sum()
```

```python
merged_table['amount']=merged_table['amount'].fillna(merged_table['amount'].mean())
```

```python
merged_table.isnull().sum()
```

```python
merged_table['email']=merged_table['email'].fillna(merged_table['name'].str.replace(' ','').str.lower()+'@unknown.com')
     
```

```python
merged_table.isnull().sum()
```

```python
merged_table['phone']=merged_table['phone'].fillna('xxx-xxx-xxxx')
merged_table['phone']=merged_table['phone'].str.replace('-','')
merged_table.loc[merged_table['phone'].str.len()<10,'phone']='xxxxxxxxx'
```

```python
merged_table['phone']=merged_table['phone'].str.replace(r'\D','',regex=True)
merged_table['phone']=merged_table['phone'].str.slice(0,11)
```

```python
merged_table.isnull().sum()
```

```python
merged_table
```

```python
high_value_transactions=merged_table[merged_table['amount']>1000]
```
